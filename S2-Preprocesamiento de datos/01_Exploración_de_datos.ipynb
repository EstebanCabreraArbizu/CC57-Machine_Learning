```python
ss.pearsonr(data['AMT_INCOME_TOTAL'],data['CNT_FAM_MEMBERS'])
```

```python
#Como el pvalue > 5%, no se rechaza Ho, al 95% de confianza se concluye que NO EXISTE CORRELACIÓN entre
#los ingresos anuales de los clientes y la cantidad de miembros de familia
ss.spearmanr(data['AMT_INCOME_TOTAL'],data['CNT_FAM_MEMBERS'])
```

```python
#Prueba Chi-Cuadrado
#Ho: no existe dependencia entre las variables
#Ha: existe dependencia entre las variables
```

```python
#Tabla de contingencia
contigencia = pd.crosstab(data['FLAG_OWN_CAR'], data['CODE_GENDER'])
contigencia
```

```python
#Como el pvalue < 5%, se rechaza Ho, al 95% de confianza se concluye que EXISTE DEPENDENCIA entre
#FLAG_OWN_CAR y CODE_GENDER
c, p, dof, expected = chi2_contingency(contigencia)
print(p)
```
